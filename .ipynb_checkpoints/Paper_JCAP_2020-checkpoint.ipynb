# A low Hubble Constant from galaxy distribution observations

$\textbf{ABSTRACT:}$  This code is an example of a statistical analysis that aims to obtain new and tight estimates on Hubble  parameter, obtained by considering two data sets from galaxy distribution observations: galaxy cluster gas mass fractions and baryon acoustic oscillation measurements. Here I detail a Python version (the original was written in the Python and Wolfram language) of the code used to obtain the results published in the renowned Journal of Cosmology and Astroparticle: __[A low Hubble constant from galaxy distribution observations](\href{https://dx.doi.org/10.1088/1475-7516/2020/09/053)__. In order to facilitate the understanding of the code, we have inserted some important parts of that article.

## Cosmological Models and data sets.

```python
### Setting the value of fixed parameters ###
velc = 299792.4580  # [km/s]
T0cmb = 2.725  # [K] Fixsen 2009 DOI:10.1088/0004-637X/707/2/916
Ogh2 = (4.482*10**(-7.0))*T0cmb**4.0 # Omega_gamma x h^2 (eq.8 arxiv:1411.1074)
Neff = 3.046  # Num. efetivo de família de neltrinos.
Orh2 = Ogh2*(1.0+0.2271*Neff)  # Omega_rad x h^2 (fontons + neutrinos)
```

The Hubble parameter in the **Non-flat XCDM model** [km/s/Mpc]:}

$$ H(z)=H_0 E(z)=100 h \sqrt{\Omega_m (1+z)^3 +\Omega_k (1+z)^2 +\Omega_r (1+z)^4+(1-\Omega_{tot} )(1+z)^{-3(1+w_x)}} $$

* When $\Omega_k=0$ is fixed, we recover the $H(z)$ of the **Flat XCDM model**.

* When $w_x=-1$ is fixed, we recover the $H(z)$ of the **Non-flat $\Lambda$CDM model**.

* When $\Omega_k=0$ and $w_x=-1$ is fixed, we recover the $H(z)$ of the **Flat $\Lambda$CDM model**.

```python
def Hzth(oc, ob, ok, h, wx, z):
    om = oc + ob
    Otot = om+ok+(Orh2/h**2.0)
    H0 = 100*h  # (km/s)/Mpc
    Ez = (om*(1+z)**3.0 + ok*(1+z)**2.0 + (Orh2/h**2)*(1+z)**4.0
          + (1.0-Otot)*(1+z)**(3+3*wx))**(1/2)
    return H0*Ez
```

The relevant physical quantities for the BAO data are the angular diameter distance:
\begin{equation}
D_{\text{A}}(z) = \frac{1}{(1+z)}\times %\\
\left\{ \begin{array}{lll} \frac{H_{0}^{-1}}{\sqrt{|-\Omega_{\text{k,0}}|}}\text{sin}\left(\frac{\sqrt{|-\Omega_{\text{k,0}}|}}{H_{0}^{-1}}\int_{0}^{z}\frac{dz'}{H(z')} \right) & \text{if} &  \Omega_{\text{k,0}}<0  \\
									\int_{0}^{z}\frac{dz'}{H(z')} & \text{if} &  \Omega_{\text{k,0}}= 0  \\
                                    \frac{H_{0}^{-1}}{\sqrt{|-\Omega_{\text{k,0}}|}}\text{sinh}\left(\frac{\sqrt{|-\Omega_{\text{k,0}}|}}{H_{0}^{-1}}\int_{0}^{z}\frac{dz'}{H(z')} \right) & \text{if} &  \Omega_{\text{k,0}}>0  
\end{array} \right . \ ,
\label{DA z}
\end{equation}
the spherically-averaged distance:
\begin{equation}
    D_{\text{V}}(z)=\left[(1+z)^{2}D^{2}_{\text{A}}(z)\frac{z}{H(z)} \right]^{1/3}
\end{equation}
and comoving sound horizon at drag redshift em [Mpc] (see D. J. Eisenstein and W. Hu and 1998 Percival et al.: 0907.1660 Section 6):
\begin{equation}
    r_{\text{s}}(z_{\text{d}})=\int_{0}^{1/(1+z_d)} \frac{c_s(a)}{a^2H(a)}da = \frac{2}{3k_{\text{eq}}}\sqrt{\frac{6}{R(z_{\text{eq}})}}\text{ln}\left[\frac{\sqrt{1+R(z_{\text{d}})}+\sqrt{R(z_{\text{d}})+R(z_{\text{eq}})}}{1+\sqrt{R(z_{\text{eq}})}}\right] ,
\end{equation}
where $c_s(a)=c/\sqrt{3\left(1+R(a)\right)}$ is the sound horizon at the drag epoch,  $z_{\text{d}}$ is the drag epoch redshift, $z_{\text{eq}}$ is the equality redshift, $k_{\text{eq}}$ is the scale of the particle horizon at the equality epoch and \citep{Eisenstein_1998}
\begin{equation}
    R(z)\equiv \frac{3\rho_{\text{b}}}{4\rho_{\gamma}}=\frac{3\Omega_bh^2}{4\Omega_\gamma h^2}a=31.5(\Omega_{\text{b}}h^{2})\left(\frac{T_{\text{CMB}}}{2.7 \ \text{K}}\right)^{-4}\left(\frac{z}{10^3}\right)^{-1}
\end{equation} 
is the ratio of the baryon to photon momentum density. Here we determine $z_{\text{eq}}$, $k_{\text{eq}}$ and $z_{\text{d}}$, using the fit obtained by \cite{Eisenstein_1998}.

<hr>

**_The fit  of  the $z_{eq}$, $k_{eq}$ and $z_{d}$ parameters:_**

The value of $z_{\text{eq}}$, $k_{\text{eq}}$ and $z_{\text{d}}$ are obtained through formula fit obtained by Daniel J. Eisenstein and Wayne Hu (1997) (see [**arxiv: 9709112**](https://arxiv.org/abs/astro-ph/9709112) ):

$$ z_{\text{eq}}=\frac{\Omega_m h^2}{\Omega_r h^2}-1=  2.5\cdot 10^4 (\Omega_{\text{m}}h^{2}) \left(\frac{T_{\text{CMB}}}{2.7 \ \text{K}}\right)^{-4} \ ,$$ 

$$ k_{\text{eq}} =\sqrt{ 2\Omega_m H_{0}^{2}z_{\text{eq}}} \ ,$$

and

   $$ z_d = {\color{blue}{1291}}\left[\frac{(\Omega_mh^2)^{0.251}}{1+0.659(\Omega_mh^2)^{0.828}}\right]\left[1+b_1(\Omega_{b}h^2)^{b_2}\right] \ ,$$
   where
$$b_1 = 0.313 (\Omega_m h^2)^{-0.419}\left[1+0.607 (\Omega_m h^2)^{0.674}\right] \ \ \ \text{e} \ \ \ b_2 = 0.238(\Omega_m h^2)^{0.223}  \ . $$


  1. **_Note_**: Later I noticed that some authors use another version of this fit, i.s. the version presented by Wayne Hu and Naoshi Sugiyama (1996) (see: [**arxiv: 9510117**](https://arxiv.org/abs/astro-ph/9510117) , Appendix E, Eq.(E-2)), given by:
   $$ z_d = {\color{blue}{1345}}\left[\frac{(\Omega_mh^2)^{0.251}}{1+0.659(\Omega_mh^2)^{0.828}}\right]\left[1+b_1(\Omega_{b}h^2)^{b_2}\right] \ .$$ 
   where
      $$b_1 = 0.313 (\Omega_m h^2)^{-0.419}\left[1+0.607 (\Omega_m h^2)^{0.674}\right] \ \ \ \text{e} \ \ \ b_2 = 0.238(\Omega_m h^2)^{0.223}  \ , $$
   **_Note that the only difference is the number highlighted in blue, so I believe it was a typo by Daniel J. Eisenstein and Wayne Hu. This is important because the choice of one or the other may influence parameter values, including $H_0$_**.
   
2. **_Note_**: In the formulas presented in this work (in the markdown-cells), we used the notation $c=1$ (speed of light equal to 1), however, in the code, we incited it in order to recover the correct dimension of the magnitudes and their respective values.

```python
def DAth(oc, ob, ok, h, wx, z):
    # Integral:
    def integrando_Xth(zz, oc, ob, ok, h, wx): return velc / \
        Hzth(oc, ob, ok, h, wx, zz)
    resut_Xth, erro_Xth = si.quad(
        integrando_Xth, 0, z, args=(oc, ob, ok, h, wx))

    # remember that: H_0 = 100h (Km/s)/Mpc and [c/H_0] => Mpc
    if ok < 0:
        return (1.0/(1.0+z))*(velc/(100.0*h*np.sqrt(np.abs(-ok))))\
            * np.sin((100.0*h*np.sqrt(np.abs(-ok))/velc)*resut_Xth)
    elif ok > 0:
        return (1.0/(1.0+z))*(velc/(100.0*h*np.sqrt(np.abs(-ok))))\
            * np.sinh((100.0*h*np.sqrt(np.abs(-ok))/velc)*resut_Xth)
    else:
        return (1.0/(1.0+z))*resut_Xth
    
def DVth(oc, ob, ok, h, wx, z):
    return ((((1.0+z)*DAth(oc, ob, ok, h, wx, z))**2.0)*velc \
            * z/Hzth(oc, ob, ok, h, wx, z))**(1.0/3.0)

###  In integral form: ##################
# def cs(a, ob, h):
#     Rz = (3.0*ob*h*h/(4.0*Ogh2))*a
#     return velc/(3.0*(1.0+Rz))**(1.0/2.0)

# def rsth(oc, ob, ok, h, wx, z):
#     def integrando_rs(a, oc, ob, ok, h, wx): return cs(
#         a, ob, h)/((a**2.0)*Hzth(oc, ob, ok, h, wx, 1.0/a - 1.0))
#     resut_rs, erro_rs = si.quad(
#         integrando_rs, 0, 1.0/(1.0+z), args=(oc, ob, ok, h, wx))
#     return resut_rs
########################################

# zd  (see eq 4 astro-ph/9709112)
def zd(oc, h, ob):
    om = oc + ob
    b1 = (0.313*(om*h**2.0)**(-0.419))*(1.0+0.607*(om*h**2.0)**0.674)
    b2 = 0.238*(om*h**2.0)**0.223
    zd_res = 1291*(((om*h**2.0)**0.251)/(1.0+0.659 * \
                   (om*h**2.0)**0.828))*(1.0+b1*(ob*h*h)**b2)
    return zd_res

# #In numerical fit form:
def rsth(oc, ob, ok, h, wx):
    om = oc+ob

    zeq = (2.5*10**4)*(om*h**2)*(T0cmb/2.7)**(-4)  # ou  zeq = om*h*h/Orh2 -1
    keq = math.sqrt((2*om*(100*h/velc)**2)*zeq)  # unid: 1/Mpc

    Rzeq = 31.5*ob*h*h*((T0cmb/2.7)**(-4))*(zeq/1000)**(-1) #(3.0*ob*h*h/(4.0*Ogh2))*(1/(1+zeq))
    Rzd = 31.5*ob*h*h*((T0cmb/2.7)**(-4))*(zd(oc, h, ob)/1000)**(-1)

    rs_res = (2/(3*keq))*math.sqrt(6/Rzeq) * \
        math.log((math.sqrt(1+Rzd)+math.sqrt(Rzd+Rzeq))/(1+math.sqrt(Rzeq)))

    return rs_res
```

For survey set I, the BAO quantity is given by:
\begin{equation}
d_z(z)=\frac{r_{\text{s}}(z_{\text{d}})}{D_{\text{V}}(z)} \,, \label{dz}
\end{equation}
with a $\chi^{2}$ function given by:
\begin{equation}
\chi^{2}_{\text{BAO,I}}=\sum_{i=1}^{2} \left[\frac{d_{z}^{\text{th}}(z_{i})-d_{z, \ i}^{\text{ob}}}{\sigma_{d_{z , \ i}^{\text{ob}}}}\right]^2 \,.   
\end{equation}

```python
z_baoI = DataBAOI["z"]  # listas dos valores de z
dz_baoI = DataBAOI["dz"]  # listas dos valores de dz
s_baoI = DataBAOI["sigma_dz"]  # listas dos valores de sigma_dz

# log_likelihood = -(1/2)Chi2
def log_likelihood_BAO_I(oc, ob, ok, h, wx):
    chi2BaoI = sum(((rsth(oc, ob, ok, h, wx)/DVth(oc, ob, ok, h, wx, \
                   z_baoI[n])-dz_baoI[n])/s_baoI[n])**2.0 for n in range(len(z_baoI)))
    return -0.5*(chi2BaoI)
```

On the other hand, the BAO quantity for survey set II is given by:
\begin{equation}
\mathcal{D}(z)=\frac{D_{\text{V}}(z)}{r_{\text{s}}(z_{\text{d}})} r_s^{\rm fid} \,, \label{alpha}
\end{equation}
and, in this specific case, the $\chi^{2}$ function is:
\begin{align}
\chi^{2}_{\text{BAO,II}} = &\sum_{i=1}^{6} \left[\frac{\mathcal{D}^{\text{th}}(z_{i})-\mathcal{D}_{i}^{\text{ob}}}{\sigma_{\mathcal{D}_{i}^{\text{ob}}}}\right]^2  \nonumber \\
&  + \left[\vec{\mathcal{D}}^{\text{th}}-\vec{\mathcal{D}}^{\text{ob}}\right]^{\text{T}}\text{C}^{-1}_{\text{WiggleZ}}\left[\vec{\mathcal{D}}^{\text{th}}-\vec{\mathcal{D}}^{\text{ob}}\right] \,,   
\end{align}
where $\text{C}^{-1}_{\text{WiggleZ}}$ is the inverse covariance matrix, whose explicit form is \citep{Kazin:2014qga}:
\begin{equation}
10^{-4}\left( 
\begin{array}{rrr}  2.17898878  & -1.11633321    & 0.46982851 \\ 
                    -1.11633321  & 1.70712004    & -0.71847155 \\
                    0.46982851   & -0.71847155   & 1.65283175 \end{array} \right).
\end{equation}
Unlike the others, the data points of the WiggleZ survey are correlated.

```python
z_baoII = DataBAOII["z"]  # listas dos valores de z
Dz_baoII = DataBAOII["Dz"]  # listas dos valores de Dz
s_baoII = DataBAOII["sigma_D"]  # listas dos valores de sigma_Dz
r_sfidII = DataBAOII["rs_sfid"]  # listas dos valores de rs_sfid

z_baoIII = DataBAOIII["z"]  # listas dos valores de z
Dz_baoIII = DataBAOIII["Dz"]  # listas dos valores de Dz
r_sfidIII = DataBAOIII["rs_sfid"]  # listas dos valores de rs_sfid

C_WiggleZ_inver = [[2.17898878, -1.11633321, 0.46982851],
                   [-1.11633321, 1.70712004, -0.71847155],
                   [0.46982851, -0.71847155, 1.65283175]]

# log_likelihood = -(1/2)Chi2
def log_likelihood_BAO_II(oc, ob, ok, h, wx):
    chi2Bao1 = sum((((DVth(oc, ob, ok, h, wx, z_baoII[n])/rsth( \
        oc, ob, ok, h, wx))*r_sfidII[n] \
        - Dz_baoII[n])/s_baoII[n])**2.0 for n in range(len(z_baoII)))

    chi2Bao2 = 0
    for i in range(len(z_baoIII)):
        for j in range(len(z_baoIII)):
            chi2Bao2 = chi2Bao2 + ((DVth(oc, ob, ok, h, wx, z_baoIII[i])/rsth( \
                oc, ob, ok, h, wx))*r_sfidIII[i]-Dz_baoIII[i])*(10**(-4))* \
                C_WiggleZ_inver[i][j]*((DVth(oc, ob, ok, h, wx, z_baoIII[j])/rsth( \
                oc, ob, ok, h, wx))*r_sfidIII[j]-Dz_baoIII[j])
            
    return -0.5*(chi2Bao1 + chi2Bao2)
```

A quantidade de fração de massa de gás para um cluster é dada por:
$$
f_{\text{gas}}^{\text{X-ray}}\left( z\right) =A(z)K(z)\gamma (z) \frac{\Omega _{\text{b}}(z)}{\Omega _{\text{m}}(z)} \left[ \frac{D_{A}^{\text{fid}}\left( z\right) }{D_{A}\left( z\right) }\right] ^{\frac{3}{2}} \ , \label{Eq fgas}
$$ 
where $K=0.96$, $\gamma=0.848$, e 
$$
    A(z)=\left[\frac{H(z)D_{A}(z)}{H^{\text{fid}}(z)D^{\text{fid}}_{A}(z)} \right]^{\eta} \ \ \text{com} \ \ \eta =0.442 \ .
$$
Assumindo $\omega _{\text{b,}0}\equiv \Omega _{\text{b,}0}h^{2}$, nós podemos rescrever a equação anterior como:
$$
f_{\text{gas}}^{\text{X-ray}}\left( z\right) = 
\frac{K \ \gamma \ \omega _{\text{b,}0}}{\Omega _{\text{m,}0}h^2} \left[\frac{H(z)D_{A}(z)}{H^{\text{fid}}(z)D^{\text{fid}}_{A}(z)} \right]^{\eta}\left[ \frac{D_{A}^{\text{fid}}\left( z\right) }{D_{A}\left(z\right) }\right] ^{\frac{3}{2}}  .
$$

```python
def Az(oc, ob, ok, h, wx, z):
    om = oc+ob
    eta = 0.442
    return (Hzth(oc, ob, ok, h, wx, z)*DAth(oc, ob, ok, h, wx, z) /
            (Hzth(0.3-ob, ob, 0, 0.7, -1, z)*DAth(0.3-ob, ob, 0, 0.7, -1, z)))**eta


def Fgasth(oc, ob, ok, h, wx, z):
    om = oc+ob
    k, gamma = 0.96, 0.848
    fgasth_resut = (k*gamma*(ob*h**2.0)/(om*h**2.0))*(Az(oc, ob, ok, h, wx, z))\
        * (DAth(0.3-ob, ob, 0, 0.7, -1, z)/DAth(oc, ob, ok, h, wx, z))**(3.0/2.0)
    return fgasth_resut
```

Therefore, for this sample, the $\chi ^{2}$ function is given by,
\begin{equation}
\chi_{\text{GMF}}^{2}=\sum_{i=1}^{40}\frac{\left[ f_{\text{gas}}^{\text{th}}(z_i)-f_{\text{gas}, \ i}^{\text{ob}}\right] ^{2}}{\sigma _{\text{tot}, \ i}^{2}}\text{ ,}
\end{equation}
with a total uncertainty given by
\begin{equation}
\sigma _{\text{tot}, \ i}^{2} = \sigma^{2}_{f_{\text{gas}}^{\text{ob}}, \ i} + \left[f_{\text{gas}}^{\text{th}}(z_i) \right]^{2} \left\{ \left( \frac{\sigma_{K} }{K}\right)^{2} + \left( \frac{\sigma_{\gamma }}{\gamma }\right)^{2} \right. \left. + \ln^2 \left[\frac{H(z_i)D_{A}(z_i)}{H^{\text{fid}}(z_i)D^{\text{fid}}_{A}(z_i)}\right]\sigma_{\eta}^{2} \right\} \,
\end{equation}
where, as presented in the previous paragraph, \{$K$,$\sigma _{K}$\}$=$\{$0.96$,$0.127$\}, \{$\gamma$,$\sigma _{\gamma }$\}$=$ \{$0.848$,$0.085$\}, and \{$\eta$,$\sigma_\eta$\} $=$\{$0.442$,$0.035$\}.

```python
# log_likelihood = -(1/2)Chi2
def log_likelihoodFgas(oc, ob, ok, h, wx):
    om = oc+ob
    k, gamma = 0.96, 0.848
    sigmak2 = (0.09)**2. + (0.09)**2.  # (*sistemático + estatístico*);
    sigmagamma2 = (0.085)**2.
    eta = 0.442
    sigmaeta2 = (0.035)**2.0

    chi2Fgas = sum(((Fgasth(oc, ob, ok, h, wx, z_fgas[n])-fgas[n])**2.0) / \
                   (s_fgas[n]**2.0 + (Fgasth(oc, ob, ok, h, wx, z_fgas[n])**2.0)*( \
                       sigmak2/k**2.0 + sigmagamma2/gamma**2.0 \
                       + ((np.log(Az(oc, ob, ok, h, wx, z_fgas[n]) \
                            ** (1/eta)))**2.0)*sigmaeta2)) for n in range(0, len(z_fgas)))
    return -0.5*chi2Fgas
```

### Prior 

The prior function encodes any previous knowledge that we have about the parameters: results from other experiments, physically acceptable ranges, etc. It is necessary that you write down priors if you’re going to use MCMC because all that MCMC does is draw samples from a probability distribution and you want that to be a probability distribution for your parameters.

In this example, we’ll use uniform (so-called “uninformative”) priors on $h$, $\Omega_c$, $\Omega_k$ and $w_x$. For example, we’ll use the following conservative prior on $h$:
$$
-2\ln \left[p(h)\right] = \left \{\begin{array}{ll}
        0 \,, & \mbox{if}\,0.3 < h < 1.1 \\
        \infty \,, & \mbox{otherwise}
    \end{array}
    \right .
$$
On the other hand, we use our knowledge of the value of $\omega_b$ ($\omega_b\equiv\Omega_b h^2 = 0.0226 \pm 0.00034$, see: __[**R. J. Cooke at. al., 2016 arXiv:1607.03900**](
https://doi.org/10.3847/0004-637X/830/2/148)__), to add a Gaussian (so-called “informative”) prior to the $\Omega_b h^2$ parameter:
$$
\chi^{2}_{\text{priors,BBN}}=-2\ln \left[p(\Omega_b h^2)\right] \propto \left[\frac{\Omega_b h^2- 0.0226}{0.00034}\right]^2
$$

```python
# implementando os priors
def log_prior(oc, ob, ok, h, wx):

    # flat priors om oc, ob, ok, h e wx
    if not 0.01 < oc < 0.8:
        return -np.inf
    if not -0.6 < ok < 0.6:
        return -np.inf
    if not 0.35 < h < 1.0:
        return -np.inf
    if not -2 < wx < 0:
        return -np.inf
    if not 0.001 < ob < 0.3:
        return -np.inf
    
    # gaussian prior on obh2
    Obh2_mu = 0.0226
    Obh2_error = 0.00034
    
    log_priorBBN = -0.5*((ob*h*h-Obh2_mu)/Obh2_error)**2
    # np.log(1.0/(np.sqrt(2*np.pi)*Obh2_error))
    return log_priorBBN
```

The statistical analysis is performed by the construction of the $\chi^{2}$ function,
\begin{equation}
\chi^{2}_{\text{tot.}}(\{\alpha\})=\chi^{2}_{\text{BAO,I}}+\chi^{2}_{\text{BAO,II}}+\chi^{2}_{\text{GMF}}+ \chi^{2}_{\text{priors,BBN}}\ .
\end{equation}
From this function, we are able to construct the likelihood distribution function, $\mathcal{L}(\{\alpha\})=Be^{-\frac{1}{2}\chi^{2}_{\text{tot.}}(\{\alpha\})}$, where $B$ is the normalization factor and $\{\alpha\}$ is the set of free parameters of the cosmological model in question, that is, 
the flat  $\Lambda$CDM and XCDM models, and the non-flat $\Lambda$CDM model.

```python
# log_probability = -(1/2)Chi2_tot
def log_probability(listparam):

    #ok = 0.0       # Flat XCDM model
    wx = -1        # Non-flat  ΛCDM model
    #ok, wx = 0, -1 # Flat  Λ CDM model
        
    # Lista da parâmetros livres: 
    #oc, ob, h, wx = listparam   # Flat XCDM model
    oc, ob, ok, h = listparam   # Non-flat  ΛCDM model
    #oc, ob, h = listparam       # Flat  ΛCDM model

    # prior:
    lp=log_prior(oc, ob, ok, h, wx)
    if not np.isfinite(lp): # se lp for um número, isfinete -> True -> lpprior = lp 
        return -np.inf      # se lp não for um numero,
    lpprior = lp                     #  isfinete -> false -> log_probability = -np.inf
    
    log_probability_tot = lpprior \
          + log_likelihoodFgas(oc, ob, ok, h, wx) \
               + log_likelihood_BAO_I(oc, ob, ok, h, wx) \
                    +log_likelihood_BAO_II(oc, ob, ok, h, wx) 

    return log_probability_tot
```

### Using the emcee
After all this setup, it’s easy to sample this distribution using emcee. We’ll start by initializing the walkers in a tiny Gaussian ball around the maximum likelihood result (I’ve found that this tends to be a pretty good initialization in most cases).

```python
#Informações para instalar do emcmc: 
#  https://emcee.readthedocs.io/en/stable/user/install/

# definir valor inicial para os parâmetros
oc_ini, ob_ini, ok_ini, h_ini, wx_ini = 0.27,0.047,0.0,0.7,-1.0

#param_ini = [oc_ini, ob_ini, h_ini, wx_ini] # Flat XCDM model
param_ini = [oc_ini, ob_ini, ok_ini, h_ini] # Non-flat  ΛCDM model
#param_ini = [oc_ini, ob_ini, h_ini] # Flat  ΛCDM model

dim = len(param_ini) 

pos = param_ini + 1e-4 * np.random.randn(32,dim)
nwalkers, ndim = pos.shape

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability)
```

Then we can proceed in two different ways, one simpler and one more elaborate.
* **First way**: run a sufficiently large number (5.000, for example) of MCMC steps.

```python
# totalsteps = 5000 

# # primeira forma de iniciar o mcmc:
# sampler.run_mcmc(pos, totalsteps, progress=True);
```

* **Second way**: Now, we’ll run the chain for up to 10,000 steps and check the autocorrelation time every 100 steps. If the chain is longer than 100 times the estimated autocorrelation time and if this estimate changed by less than 1%, we’ll consider things converged.

```python
totalsteps = 10000
coords = pos
# We'll track how the average autocorrelation time estimate changes
index = 0
autocorr = np.empty(totalsteps)

# This will be useful to testing convergence
old_tau = np.inf

# Now we'll sample for up to max_n steps
for sample in sampler.sample(coords, iterations=totalsteps, progress=True):
    # Only check convergence every 100 steps
    if sampler.iteration % 100:
        continue

    # Compute the autocorrelation time so far
    # Using tol=0 means that we'll always get an estimate even
    # if it isn't trustworthy
    tau = sampler.get_autocorr_time(tol=0)
    autocorr[index] = np.mean(tau)
    index += 1

    # Check convergence
    converged = np.all(tau * 100 < sampler.iteration)
    converged &= np.all(np.abs(old_tau - tau) / tau < 0.01)
    if converged:
        break
    old_tau = tau
```

A test of whether or not the sampling went well is to check the mean acceptance fraction of the ensemble using the EnsembleSampler.acceptance_fraction() property. This number should be between approximately 0.25 and 0.5 if everything went as planned.

```python
print("Mean acceptance fraction: {0:.3f}".format(np.mean(sampler.acceptance_fraction)))
```

Let’s take a look at what the sampler has done. A good first step is to look at the time series of the parameters in the chain. The samples can be accessed using the EnsembleSampler.get_chain() method. This will return an array with the shape ("totalsteps", 32, "dim") giving the parameter values for each walker at each step in the chain. The figure below shows the positions of each walker as a function of the number of steps in the chain:

```python
fig, axes = plt.subplots(ndim, figsize=(10, 7), sharex=True)
samples = sampler.get_chain()

# Flat XCDM model
#labels = ["$\Omega_c$","$\Omega_b$","$h$","$w_x$"] 
# Non-flat  ΛCDM model
labels = ["$\Omega_c$","$\Omega_b$","$\Omega_k$","$h$"] 
# Flat  ΛCDM model
#labels = ["$\Omega_c$","$\Omega_b$","$h$"] 

ndim = len(labels)
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```

As mentioned above, the walkers start in small distributions around the values initially assigned and then they quickly wander and start exploring the full posterior distribution. In fact, after some steps, the samples seem pretty well “burnt-in”. That is a hard statement to make quantitatively, but we can look at an estimate of the integrated autocorrelation time "tau = sampler.get_autocorr_time()" (see the Autocorrelation analysis & convergence tutorial for more details).

This suggests that some steps are needed for the chain to “forget” where it started. It’s not unreasonable to throw away a few times this number of steps as “burn-in”. Let’s discard the initial 3 x tau steps, thin by about half the autocorrelation time (tau/2 steps), and flatten the chain so that we have a flat list of samples:

```python
tau = sampler.get_autocorr_time()
ndim = dim
burnin = int(3 * np.max(tau)) # discartando 3 x o valor de tau das etapas iniciais 
thin = int(0.5 * np.min(tau)) # diluir cerca da metade do tau

samples = sampler.get_chain(discard=burnin, flat=True, thin=thin)
log_prob_samples = sampler.get_log_prob(discard=burnin, flat=True, thin=thin)

print("autocorrelation time: {0}".format(tau))
print("burn-in: {0}".format(burnin))
print("thin: {0}".format(thin))
print("flat chain shape: {0}".format(samples.shape))
print("flat log prob shape: {0}".format(log_prob_samples.shape))
```

The code that follows generates a simple plot of the MCMC chain using the Getdist package, so we can check if everything is right before saving the file.

```python
sample_teste = MCSamples(samples=samples)
# # # Triangle plot
g = plots.get_subplot_plotter(width_inch=6)
g.triangle_plot([sample_teste],shaded=True,filled=False,
                    line_args=[{'lw':2,'color':"darkblue"}])
```

#### Flat $\Lambda$CDM model

We will first read the previously generated flat $\Lambda$CDM model MCMC files and then add to this array a column of the derived parameter $\Omega_m$, using the fact that $\Omega_m = \Omega_c + \Omega_b$.

```python
# Lendo os arquivos MCMC - Flat-LCDM
MCMC_BAO = pd.read_fwf("./ResultMCMC/MCMC_param_FLCDM_BAO.txt",sep=' ', names=['oc',"ob",'h'])
MCMC_Fgas = pd.read_fwf("./ResultMCMC/MCMC_param_FLCDM_Fgas.txt",sep=' ', names=['oc',"ob",'h'])
MCMC_full = pd.read_fwf("./ResultMCMC/MCMC_param_FLCDM_full.txt",sep=' ', names=['oc',"ob",'h'])
MCMC_log_prob_full = pd.read_fwf("./ResultMCMC/MCMC_log_prob_FLCDM_full.txt",
                                 sep=' ', names=['log_prop'])

# adicionando um paramentro derivado, no caso o Om
MCMC_full = MCMC_full.eval("om = oc + ob")
MCMC_Fgas = MCMC_Fgas.eval("om = oc + ob")
MCMC_BAO = MCMC_BAO.eval("om = oc + ob")
```

Calculating the mean and $1\sigma$ statistical confidence interval ($68\%$) of the parameters:

```python
from IPython.display import display, Math

sample = np.array(MCMC_full)
labels = ["\Omega_c","\Omega_b","h","\Omega_m"]
ndim = sample.shape[1]

print(" Mean and statistical confidence interval:")
alpha = (100 - 68.3)/2 # 1 sigma c.l.
#alpha = (100 - 95.4)/2 # 2 sigma c.l.
#alpha = (100 - 99.7)/2 # 3 sigma c.l.
MeanParam =[]
for i in range(ndim):
    mcmc1s = np.percentile(sample[:, i], [alpha, 50, 100 - alpha]) # 1sigmas 68.3%
    q1s = np.diff(mcmc1s)
    txt = "\mathrm{{{3}}} = {0:.3f}_{{-{1:.3f}}}^{{+{2:.3f}}}"
    txt = txt.format(mcmc1s[1], q1s[0], q1s[1], labels[i])
    MeanParam.append( np.around(mcmc1s[1],3) )
    display(Math(txt))

print(" Best fit and lowest chi-square:")
list_chi2_df = -2*MCMC_log_prob_full  # chi2
indxmin = list_chi2_df['log_prop'].idxmin() # localizando indx do valor de menor chi2
display(MCMC_full.loc[indxmin]) # mostrando valores dos parâmetros que minimiza o chi2
print("chi2min = ",np.around(list_chi2_df.loc[indxmin][0],3))
```
